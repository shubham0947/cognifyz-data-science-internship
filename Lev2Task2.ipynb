```python
import pandas as pd

df = pd.read_csv("Dataset .csv")

df.head()
```

```python
most_common_price = df['Price range'].value_counts()

print("Most common price range:")
print(most_common_price)
```

```python
print("Most common price range:", df['Price range'].mode()[0])
```

```python
avg_rating_price = df.groupby('Price range')['Aggregate rating'].mean()

print("Average rating for each price range:")
print(avg_rating_price)
```

```python
color_rating = df.groupby('Rating color')['Aggregate rating'].mean()

print("Average rating by color:")
print(color_rating)
```

```python
print("Color with highest rating:")
print(color_rating.idxmax())
```

```python
import matplotlib.pyplot as plt

df.groupby('Price range')['Aggregate rating'].mean().plot(kind='bar')

plt.title("Average Rating by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Average Rating")

plt.show()
```
