```python
import pandas as pd

df = pd.read_csv("Dataset .csv")

df.head()
```

```python
table_booking_percent = (df['Has Table booking'].value_counts(normalize=True) * 100)

print("Percentage of restaurants offering table booking:")
print(table_booking_percent)
```

```python
online_delivery_percent = (df['Has Online delivery'].value_counts(normalize=True) * 100)

print("Percentage of restaurants offering online delivery:")
print(online_delivery_percent)
```

```python
table_booking_rating = df.groupby('Has Table booking')['Aggregate rating'].mean()

print("Average ratings based on table booking:")
print(table_booking_rating)
```

```python
delivery_price = df.groupby('Price range')['Has Online delivery'].value_counts()

print(delivery_price)
```

```python
import matplotlib.pyplot as plt

df.groupby('Has Table booking')['Aggregate rating'].mean().plot(kind='bar')

plt.title("Average Rating vs Table Booking")
plt.xlabel("Table Booking")
plt.ylabel("Average Rating")

plt.show()
```
